```python
class Grocery_Item:
    def __init__(self, name, price, discount):
        self.name = name
        self.price = price
        self.has_discount = discount

    def display_info(self):
        print(self.name, "is €", self.price)

apple = Grocery_Item('apple', 1, False)
cheerios = Grocery_Item('cheerios', 4, True)
print('does', apple.name, 'have a discount?',apple.has_discount)
print('does', cheerios.name, 'have a discount?',cheerios.has_discount)
print(cheerios.name, cheerios.price, cheerios.has_discount)
print(cheerios.name, 'is €', cheerios.price)
```
